# Mapping with STAR

````python
#download genome
# Download C. elegans genome from Ensembl Metazoa release 62
!mkdir -p genome
!wget -P genome https://ftp.ebi.ac.uk/ensemblgenomes/pub/release-62/metazoa/fasta/caenorhabditis_elegans/dna/Caenorhabditis_elegans.WBcel235.dna.toplevel.fa.gz
!gunzip genome/Caenorhabditis_elegans.WBcel235.dna.toplevel.fa.gz

# Now navigate into the genome folder
%cd genome

#rename the file to c_elegans.fa
!mv Caenorhabditis_elegans.WBcel235.dna.toplevel.fa c_elegans.fa
````

````python
!mkdir -p mapped
````

````python
#create genome index
!STAR --runMode genomeGenerate --genomeDir genomeIndex --genomeFastaFiles c_elegans.fa
````

````python
%%bash
for infile in trim/*.fq; do
  outfile=$(basename "$infile" .fq)
  STAR --genomeDir genomeIndex \
       --readFilesIn "$infile" \
       --outFileNamePrefix mapped/${outfile}_ \
       --outSAMtype BAM SortedByCoordinate \
       --outSAMattributes All
done
````
